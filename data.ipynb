```python
from bs4 import BeautifulSoup as bsoup
import requests
import numpy as np
import pandas as pd

url = "https://ncc.gov.ng/technical-regulation/standards/type-approval#approved-equipment"

response = requests.get(url)
response
```

```python
soup = bsoup(response.text, 'html')

# soup
```

```python
table = soup.find_all('tbody')

info = []

for tab in table:
    rows = tab.find_all('tr')
    for row in rows:
        val = row.find_all('td')
        striped = [r.text.strip() for r in val]
        if len(striped) == 6:
            info.append(striped)
            
info
```

```python
info1 = info[:2257]
info2 = info[2257:]

(len(info1)+ len(info2)) == len(info)
```

```python
head = soup.find_all('thead')

column = []
for col in head:

    row = col.find_all('tr')
    for cols in row:
        ite = cols.find_all('td')
        data = [a.text.strip() for a in ite]
        print(data)
        column.append(data)

column
```

```python
column = np.array(column)

data = np.array(info)
column
```

```python
final_df1 = pd.DataFrame(info1, columns=column[0])
final_df2 = pd.DataFrame(info2, columns=column[0])

final_df1
```

```python
final_df1.to_csv('Approved_handset.csv', index=False)
final_df2.to_csv('Approved_equipment.csv', index=False)
```
